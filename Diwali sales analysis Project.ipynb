```python
# To know count according to Product id
sns.countplot(x='Product_ID', data=df)
```

```python
# To know count according to gender
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# To know count of age group
sns.countplot(x='Age Group', data= df)
```

```python
sales=df.groupby(by=['Gender'])['Amount'].sum().sort_values(by='Amount', ascending=False)
sales
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gen
```

```python
ax1=sns.barplot(x='Gender', y='Amount', data=sales_gen)
for bars in ax1.containers:
    ax1.bar_label(bars)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

```python
ax2=sns.countplot(x='Age Group', data= df)
for bars in ax2.containers:
    ax2.bar_label(bars)
```

```python
ax2=sns.countplot(x='Age Group', data= df, hue='Gender')
for bars in ax2.containers:
    ax2.bar_label(bars)
```

```python
# age group Vs Amount
sales_Age=df.groupby(by=['Age Group'],as_index=False)['Amount'].sum().sort_values(by=['Age Group'], ascending=True )
sales_Age
```

```python
sns.barplot(x='Age Group',y='Amount', data= sales_Age)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

```python
# Total no. of orders from top 10 buyers states
order_states= df.groupby(by=['State'],as_index=False)['Orders'].sum().sort_values(by=['Orders'], ascending= False).head(10)
order_states
```

```python
plt.figure(figsize=(25,10))
sns.barplot(data=order_states, x='State', y='Orders')
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

```python
# Total Amount from top 10 buyers states
sales_states= df.groupby(by=['State','Gender'],as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending= False)
sales_states
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x='State',y='Amount',data= sales_states, hue='Gender')
```

*From above graphs we can see that most of the orders & total sales/amount are FEMALES from Uttar Pradesh, Maharashtra and Karnataka respectively. and no males are from Andhra Pradesh, MAdhya pradesh *

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

*From above graphs we can see that most of the buyers are married and they  (women)have high purchasing power*

```python
# To know the sale according to product category
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_product, x = 'Product_Category',y= 'Amount')
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*
